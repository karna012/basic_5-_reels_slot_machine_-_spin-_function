```python
#basic 5 reels slot machine


import random

# Define the symbols and their payouts
symbols = ['Cherry', 'Lemon', 'Orange', 'Plum', 'Bell', 'Bar', '7', 'Wild']
payouts = {'Cherry': 2, 'Lemon': 3, 'Orange': 4, 'Plum': 5, 'Bell': 10, 'Bar': 20, '7': 100, 'Wild': 0}

# Define the reel strips
reel1 = ['Cherry', 'Lemon', 'Orange', 'Plum', 'Bell', 'Bar', '7', 'Wild']
reel2 = ['Cherry', 'Lemon', 'Orange', 'Plum', 'Bell', 'Bar', '7', 'Wild']
reel3 = ['Cherry', 'Lemon', 'Orange', 'Plum', 'Bell', 'Bar', '7', 'Wild']
reel4 = ['Cherry', 'Lemon', 'Orange', 'Plum', 'Bell', 'Bar', '7', 'Wild']
reel5 = ['Cherry', 'Lemon', 'Orange', 'Plum', 'Bell', 'Bar', '7', 'Wild']

# Define the player's starting balance and bet amount
balance = 100
bet = 1

# Define the wild symbol's multiplier and bonus feature
wild_multiplier = 2
bonus_trigger = 3

# Function to spin the reels and determine the payout
def spin():
    global balance
    global bet
    # Subtract the bet from the balance
    balance -= bet
    # Spin the reels
    reel1_result = random.choice(reel1)
    reel2_result = random.choice(reel2)
    reel3_result = random.choice(reel3)
    reel4_result = random.choice(reel4)
    reel5_result = random.choice(reel5)
    # Determine the payout based on the results
    symbols = [reel1_result, reel2_result, reel3_result, reel4_result, reel5_result]
    payout = 0
    for symbol in set(symbols):
        if symbols.count(symbol) >= 3:
            if symbol == 'Wild':
                payout += payouts[symbol] * wild_multiplier * bet
            else:
                payout += payouts[symbol] * bet
    if symbols.count('Wild') == bonus_trigger:
        balance += 10 * bet
    # Cap the maximum win at 2500X
    if payout > 2500 * bet:
        payout = 2500 * bet
    # Add the payout to the balance
    balance += payout
    # Return the results
    return (reel1_result, reel2_result, reel3_result, reel4_result, reel5_result, payout)

# Play the game
print('5x5 Slot Machine')
print('Starting balance:', balance)
while balance >= bet:
    input('Press Enter to spin the reels...')
    result = spin()
    print('Reel 1:', result[0])
    print('Reel 2:', result[1])
    print('Reel 3:', result[2])
    print('Reel 4:', result[3])
    print('Reel 5:', result[4])
    if result[5] > 0:
        print('You won', result[5])
    else:
        print('Sorry, you didn\'t win')
    print('Balance:', balance)
print('Game over')
```
